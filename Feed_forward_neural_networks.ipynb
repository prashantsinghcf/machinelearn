```
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt
```

```
dataset = loadtxt("/content/pima-indians-diabetes.data.csv.txt",delimiter=',')
X = dataset[:,:8]
Y = dataset[:,8]

model=Sequential()
model.add(Dense(64, input_dim=8, activation='relu'))
model.add(Dense(16,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

history = model.fit(X, Y, validation_split=0.33, epochs=150, batch_size=30, verbose=0)

plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title("Model Accuracy")
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Train','Test'],loc='upper left')
plt.show()

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title("Model Loss")
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Train','Test'],loc='upper left')
plt.show()

print('Train Accuracy: ',history.history['acc'][-1])
print('Test Accuracy: ',history.history['val_acc'][-1])
```
